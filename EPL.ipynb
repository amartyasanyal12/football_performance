```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(df['HomeTeam'][df['FTR']=='H'],data=df)
plt.title("Most wins at Home",fontsize=19)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(df['HomeTeam'][df['FTR']=='A'],data=df)
plt.title("Most Losses at Home",fontsize=19)
plt.xticks(rotation=90)
plt.show()
```

```python
temp=pd.DataFrame({"Final results 09-08 season":df['FTR']})
count_wins=temp.value_counts()
print(count_wins)
```

```python
teams=df["HomeTeam"].unique()
team_performance=pd.DataFrame([],columns=['Home Team','No.of Games Won at Home'])
for id,element in enumerate(teams):
    temp_team=df[(df['HomeTeam']==element) & (df['FTR']=='H')]
    team_performance=team_performance.append({'Home Team':element,
                                            'No.of Games Won at Home':temp_team['HomeTeam'].count()},ignore_index=True)
```

```python
team_performance
```

```python
teams=df["HomeTeam"].unique()
team_performance2=pd.DataFrame([],columns=['Home Team','No.of Games Lost at Home'])
for id,element in enumerate(teams):
    temp_team1=df[(df['HomeTeam']==element) & (df['FTR']=='A')]
    team_performance2=team_performance2.append({'Home Team':element,
                                            'No.of Games Lost at Home':temp_team1['HomeTeam'].count()},ignore_index=True)
```

```python
team_performance2
```

```python
sns.catplot(x="Home Team", y="No.of Games Lost at Home", data=team_performance2, kind='bar', height=3,aspect=2)
plt.xticks(rotation=90)
plt.xlabel("EPL Teams")
plt.ylabel("No. of Games Lost at home")
```

```python
sns.catplot(x="Home Team", y="No.of Games Lost at Home", data=team_performance2, kind='bar', height=3,aspect=2)
plt.xticks(rotation=90)
plt.xlabel("EPL Teams")
plt.ylabel("No. of Games Lost at home")
```
